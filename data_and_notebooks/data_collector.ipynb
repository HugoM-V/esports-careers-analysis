```python
# Collect data and create players_profiles.csv & players_tournaments.csv
import requests
import urllib3
import pandas as pd
import time
from pathlib import Path

# Disable SSL warning
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)

# Constants
API_KEY = '262234342ed4514a8b4c9ac73e726504e22d609cba266b710eb17249c45bd02a'
BASE_URL = 'http://api.esportsearnings.com/v0'
CSV_PROFILES = 'players_profiles.csv'
CSV_TOURNAMENTS = 'players_tournaments.csv'
SKIPPED_LOG = 'skipped_players.txt'

# Load full player list
df_players = pd.read_csv("./top_1000_players.csv")

# Choose the batch
start_index = 800  # üîÅ change to 200, 400, 600... for next batches
batch_size = 200
end_index = start_index + batch_size
df_batch = df_players.iloc[start_index:end_index]

# Storage for batch data
profile_data = []
tournament_data = []

# API helper functions
def get_player_profile(player_id):
    url = f"{BASE_URL}/LookupPlayerById?apikey={API_KEY}&playerid={player_id}"
    response = requests.get(url, verify=False)
    return response.json() if response.status_code == 200 else None

def get_player_tournaments(player_id):
    url = f"{BASE_URL}/LookupPlayerTournaments?apikey={API_KEY}&playerid={player_id}"
    response = requests.get(url, verify=False)
    return response.json() if response.status_code == 200 else []

# Loop through batch
for _, row in df_batch.iterrows():
    player_id = row["PlayerId"]
    print(f"Fetching Player ID: {player_id}")
    
    try:
        profile = get_player_profile(player_id)
        if profile:
            profile_data.append(profile)
        
        tournaments = get_player_tournaments(player_id)

        if isinstance(tournaments, str):
            print(f"‚ö†Ô∏è No tournaments for player {player_id}")
            continue
        
        for t in tournaments:
            t["PlayerId"] = player_id
            tournament_data.append(t)

        time.sleep(1)  # polite delay

    except Exception as e:
        print(f"‚ùå Error for player {player_id}: {e}")
        with open(SKIPPED_LOG, "a") as f:
            f.write(f"{player_id}\n")
        continue

# Convert to DataFrames
df_profiles_batch = pd.DataFrame(profile_data)
df_tournaments_batch = pd.DataFrame(tournament_data)

# Save or append to CSV
first_batch = not Path(CSV_PROFILES).exists()

df_profiles_batch.to_csv(CSV_PROFILES, mode='a', index=False, header=first_batch)
df_tournaments_batch.to_csv(CSV_TOURNAMENTS, mode='a', index=False, header=first_batch)

print(f"‚úÖ Batch {start_index}-{end_index - 1} saved.")
```

```python
# Collect game data based on GameID from players_tournament.csv and create games_metadata.csv

import pandas as pd
import requests
import urllib3
from tqdm import tqdm
import time
import os
from dotenv import load_dotenv

# Disable SSL warnings
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)

# Load API key from .env file
load_dotenv()
API_KEY = os.getenv("API_KEY")

# Load players_tournaments.csv
df = pd.read_csv("players_tournaments.csv")

# Extract unique GameIds
unique_game_ids = df["GameId"].dropna().unique()

# Store results
games_metadata = []

# Fetch game metadata for each GameId
for game_id in tqdm(unique_game_ids, desc="Collecting game data"):
    url = f"http://api.esportsearnings.com/v0/LookupGameById?apikey={API_KEY}&gameid={int(game_id)}"
    try:
        response = requests.get(url, verify=False, timeout=10)
        if response.status_code == 200:
            data = response.json()
            data["GameId"] = game_id  # Re-add GameId for reference
            games_metadata.append(data)
        else:
            print(f"‚ö†Ô∏è Error for GameId {game_id}: {response.status_code}")
    except Exception as e:
        print(f"‚ùå Exception for GameId {game_id}: {e}")
    time.sleep(1.2)  # Respect API rate limits

# Save to CSV
df_games = pd.DataFrame(games_metadata)
df_games.to_csv("games_metadata.csv", index=False)
print("‚úÖ Saved as games_metadata.csv")
```

```python
# Script to extend tournaments for players stuck at 100 results
import requests
import urllib3
import pandas as pd
import time
from pathlib import Path
import os
from dotenv import load_dotenv

# Disable SSL warning
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)

# Load API key from .env file
load_dotenv()
API_KEY = os.getenv("API_KEY")

# Constants
BASE_URL = 'http://api.esportsearnings.com/v0'
INPUT_PLAYER_IDS = 'incomplete_player_ids.csv'
EXTENDED_TOURNAMENTS_CSV = 'players_tournaments_extended.csv'
SKIPPED_LOG = 'skipped_players_extended.txt'

# Load list of incomplete PlayerIds
df_ids = pd.read_csv(INPUT_PLAYER_IDS)

# Storage for fetched tournaments
tournament_data = []

# API helper function for paginated tournament retrieval
def get_all_tournaments(player_id):
    all_tournaments = []
    offset = 0
    while True:
        url = f"{BASE_URL}/LookupPlayerTournaments?apikey={API_KEY}&playerid={player_id}&offset={offset}"
        response = requests.get(url, verify=False)

        if response.status_code != 200:
            break

        data = response.json()

        if not data or isinstance(data, str):  # empty or error message
            break

        for t in data:
            t["PlayerId"] = player_id
            all_tournaments.append(t)

        if len(data) < 100:
            break

        offset += 100
        time.sleep(1)  # polite delay

    return all_tournaments

# Loop through players and fetch extended tournament data
for _, row in df_ids.iterrows():
    player_id = row["PlayerId"]
    print(f"üîÑ Extending tournaments for Player ID: {player_id}")

    try:
        tournaments = get_all_tournaments(player_id)
        tournament_data.extend(tournaments)

    except Exception as e:
        print(f"‚ùå Error for player {player_id}: {e}")
        with open(SKIPPED_LOG, "a") as f:
            f.write(f"{player_id}\n")
        continue

# Convert and save to CSV
df_extended = pd.DataFrame(tournament_data)
df_extended.to_csv(EXTENDED_TOURNAMENTS_CSV, index=False)
print(f"‚úÖ Extended tournament data saved to {EXTENDED_TOURNAMENTS_CSV}")
```

```python
# Script to merge original and extended tournament data
import pandas as pd

# File paths
original_csv = "players_tournaments.csv"
extended_csv = "players_tournaments_extended.csv"
output_csv = "players_tournaments_merged.csv"

# Load datasets
df_original = pd.read_csv(original_csv)
df_extended = pd.read_csv(extended_csv)

# Combine all dataframes
df_combined = pd.concat([df_original, df_extended], ignore_index=True)

# Drop duplicates based on PlayerId + TournamentName + EndDate (safe subset)
df_deduplicated = df_combined.drop_duplicates(subset=["PlayerId", "TournamentName", "EndDate"])

# Save to new CSV
df_deduplicated.to_csv(output_csv, index=False)
print(f"‚úÖ Merged and deduplicated tournament data saved to {output_csv}")
```
